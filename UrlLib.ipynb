```python
import json
import urllib.request

id=int(input("enter id"))
req=urllib.request.Request(url=f'https://jsonplaceholder.typicode.com/posts/{id}')

with urllib.request.urlopen(req) as resp:
    data=json.loads(resp.read().decode('utf-8'))

print(data)
print(type(data))
print(data['title'])
```

```python
import json
import urllib.request

data={
  "userId": 1,
  "title": "CGI Adv Python Training",
  "body": "CGI Adv Python Training by Abhishek Samanta"
}
print(data)
data=json.dumps(data)

req=urllib.request.Request(url='https://jsonplaceholder.typicode.com/posts/',data=bytes(data.encode("utf-8")),method="POST")
req.add_header("Content-type","application/json; charset = UTF-8")

with urllib.request.urlopen(req) as resp:
    print(resp.getcode())
    print(resp.status)
    response=json.loads(resp.read().decode('utf-8'))
    print(response)
```
